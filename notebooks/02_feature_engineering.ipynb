```python
import pandas as pd

df = pd.read_parquet("../data/raw_data_full.parquet")
df["datetime"] = pd.to_datetime(df["datetime"])
print(df.head())
print(df.dtypes)
```

```python
df["lag_1h"] = df["cnt"].shift(1)
df["lag_24h"] = df["cnt"].shift(24)
df["lag_168h"] = df["cnt"].shift(168)
df.head(10)
df = df.dropna()
df.head(10)
```

```python
df["rolling_mean_3h"] = df["cnt"].rolling(window=3).mean()
df["rolling_mean_24h"] = df["cnt"].rolling(window=24).mean()
df = df.dropna()
print(df.shape)
df.head(10)
```

```python
df.to_parquet("../data/processed_data.parquet")
```
